```python
import pandas as pd
import numpy as np
import folium
import geojson
from geojson import Feature, FeatureCollection, Point
import json
```

```python
df = pd.read_csv('../livablestreets/data/data_Berlin_grid_1000m.csv')
df.head()
```

```python
df_centers = df[['lat_center','lng_center']]
df_centers['center']='center'
df_centers
```

```python
# columns used for constructing geojson object
features = df_centers.apply(
    lambda row: Feature(geometry=Point((float(row['lat_center']), float(row['lng_center'])))),
    axis=1).tolist()

# all the other columns used as properties
properties = df_centers.drop(['lat_center', 'lng_center'], axis=1).to_dict('records')

# whole geojson object
feature_collection = FeatureCollection(features=features, properties=properties)
```

```python
feature_collection
```

```python
geo_centers = folium.GeoJson(feature_collection, name="centers",show=True, zoom_on_click=True)
```

```python
mapObj = folium.Map(location=[52.5200, 13.4050], zoom_start=11)
geo_centers.add_to(mapObj)
folium.LayerControl().add_to(mapObj)
mapObj
```
